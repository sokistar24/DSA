# Introduction to Algorithm Analysis Using Big O



let compare two different alogirthms
$$ \sum_{i=0}^{n} {i}  $$

So how can we objectively compare these algorithms?

Algorithm 1: loops through every all the number and adds the sum

$$ \sum_{i=0
}^{n} {i} = 0+ 1 + 2 + 3 + \cdots + n $$


Algorithm 2: Takes the product of n and n+1 divides that by

$$ \sum_{i=0}^{n} {i} = \frac{n(n+1)}{2} $$

# Code for algorithm 1

```python

def sum1(n):
    '''
    Take an input of n and return the sum of the numbers from 0 to n
    '''
    final_sum = 0
    for x in range(n+1):
        final_sum += x

    return final_sum
```

# Code for algorithm 2

```python
def sum2(n):
    """
    Take an input of n and return the sum of the numbers from 0 to n
    """
    return (n*(n+1))/2
```

# Compare the results

```python
sum1(10)
```

```python
sum2(10)
```

You'll notice both functions have the same result, but completely different algorithms.

# compare the Big- O notations (their time complexity)

Algorithm 1

```python

def sum1(n):
    '''
    Take an input of n and return the sum of the numbers from 0 to n
    '''
    final_sum = 0
    for x in range(n+1):
        final_sum += x

    return final_sum
```

 sum1 function will create an assignment n+1 times, we can see this from the for loop. This means it will assign the final_sum variable n+1 times. We can then say that for a problem of n size (in this case just a number n) this function will take 1+n steps.

This n notation allows us to compare solutions and algorithms relative to the size of the problem,
since sum1(10) and sum1(100) would take very different times to run but be using the same algorithm.


O(n+1) the 1 is insignificant as n increases thus we have

 O(n)

Algorithm 2

```python
def sum2(n):
    """
    Take an input of n and return the sum of the numbers from 0 to n
    """
    return (n*(n+1))/2
```



$$
\text{sum}(n) = \frac{n(n+1)}{2}
$$

For example:

$$
\text{sum}(10) = \frac{10 \times 11}{2}
$$

$$
\text{sum}(100) = \frac{100 \times 101}{2}
$$

Even as \(n\) increases, the computation always consists of:
- one multiplication,
- one addition,
- one division.

So the number of operations does **not** increase with \(n\), thus the time complexity

$$
O(1)
$$


# Lets also compare them using the Timeit function

Note that this is not an **Objective** comparison as it depends on hard ware only for **illustration** **purpose**



 The [**%timeit**](https://ipython.org/ipython-doc/3/interactive/magics.html#magic-timeit)  will repeat the loop iteration a certain number of times and take the best result. Check out the link for the documentation.

Let's go ahead and compare the time it took to run the functions:

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

We can see that the second function is much more efficient! Running at a much faster rate than the first.


# Visualisation

```python
# Time the loop version
res_loop = %timeit -o sum1(100)

# Time the formula version
res_formula = %timeit -o sum2(100)
```

# Visualising both for same input size (n)

```python
import matplotlib.pyplot as plt

times = [res_loop.best, res_formula.best]
labels = ['sum1', 'sum2']

plt.figure(figsize=(5, 4))
plt.bar(labels, times, color=['steelblue', 'orange'])
plt.ylabel("Time (seconds, best of runs)")
plt.title("Timing comparison for n = 100")
plt.show()
```
